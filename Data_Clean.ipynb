```python
import pandas as pd
import numpy as np
import csv
import json
import os
from os import listdir,makedirs
from os.path import isfile, join,exists
import json
from spacy.lang.en import English
from collections import Counter,defaultdict
import sys
```

```python
DATA_HOME = "minus_hein/"
minus_hein = []

maxInt = sys.maxsize
while True:
    try:    
        csv.field_size_limit(maxInt)
        break
    except OverflowError:
        maxInt = int(maxInt/10)
    
with open(DATA_HOME+"/"+"cofea_output.csv", encoding = 'utf-8') as csvfile:
    csvreader = csv.DictReader(csvfile,delimiter = '\t')
    for row in csvreader:
        minus_hein.append(row)
```

```python
# Initialize Tokenizer
nlp = English()
tokenizer = nlp.tokenizer
```

```python
# source count
sources = [x['source'] for x in minus_hein]
sources = Counter(sources)
for key,value in sources.most_common():
    print(key + " " + str(value))
```

```python
# documents per decade
decade = [x['decade'] for x in minus_hein]
decade = Counter(decade)
total = 0
for key,value in sorted(decade.items()):
    print(key + " " + str(value))
    total += value
print("total: " + str(total))
```

```python
# tokenize
for doc in minus_hein:
    tokens = tokenizer(doc['body'])
    doc['tokens'] = tokens
    doc['token_count'] = len(tokens)
```

```python
# tokens per decade
decade_tokens = defaultdict(int)
for doc in minus_hein:
    decade_tokens[str(doc['decade'])] += doc['token_count']
    
total = 0
decade_counter = Counter(decade_tokens)
for key,value in sorted(decade_counter.items()):
    print(key + " " + str(value))
    total+= value
print("total: " + str(total))
```

```python
# Full cofea dataset
cofea_dir = "cofea_full"
cofea_files = [f for f in listdir(cofea_dir) if isfile(join(cofea_dir, f)) and ".json" in f]
cofea_full = []
for file in cofea_files:
    with open(cofea_dir+"/"+file, encoding = 'utf-8') as f:
        data = json.load(f)
        cofea_full = cofea_full + data
```

```python
# source break down
cofea_sources = [x['source'] for x in cofea_full ]
cofea_sources = Counter(cofea_sources)

for key,value in cofea_sources.most_common():
    print(key + " " + str(value))
```

```python
# Documents per decade
cofea_decade = []
for x in cofea_full:
    if 'decade' in x.keys():
        cofea_decade.append(str(x['decade']))
    else:
        cofea_decade.append("NA")
    
cofea_decade = Counter(cofea_decade)
total = 0
for key,value in sorted(cofea_decade.items()):
    print(key + " " + str(value))
    total+= value

print("total: " + str(total))
```

```python
# tokenize
for doc in cofea_full:
    tokens = tokenizer(doc['body'])
    doc['tokens'] = tokens
    doc['token_count'] = len(tokens)
```

```python
# Token count broken out by decade
decade_tokens = defaultdict(int)
for doc in cofea_full:
    if 'decade' in doc.keys():
        decade_tokens[str(doc['decade'])] += doc['token_count']
    else:
        decade_tokens['NA']+=doc['token_count']

total = 0
decade_counter = Counter(decade_tokens)
for key,value in sorted(decade_counter.items()):
    print(key + " " + str(value))
    total+= value
print('total: '+str(total))
```
